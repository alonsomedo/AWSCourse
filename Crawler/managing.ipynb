```python
import boto3
import os 
```

```python
os.environ.setdefault('AWS_PROFILE', 'alonsomedo-aws')
os.environ.setdefault('AWS_DEFAULT_REGION', 'us-east-1')
```

```python
glue_client = boto3.client('glue')
list_of_crawlers = glue_client.list_crawlers()['CrawlerNames']
crawler = list_of_crawlers[2]
```

```python
glue_client.get_crawler(
    Name = crawler
)
```

```python
glue_client.get_databases()['DatabaseList']
```

```python
for db in glue_client.get_databases()['DatabaseList']:
    print(db['Name'])
```

```python
glue_client.get_tables(
    DatabaseName = 'retail_db'
)['TableList']
```

```python
for table in glue_client.get_tables(
    DatabaseName = 'retail_db'
)['TableList']:
    print(table['Name'])
```
